# 06 - Update Games

Scan ROM folders, match against a video game metadata dataset, and build/update `gamelist.xml` files.

**Pipeline overview:**
1. Scan game folders to build an inventory of owned ROMs
2. Load a cleaned metadata dataset (`.pkl`)
3. Build or update `gamelist.xml` files using one of three source modes:
   - `files_only` — inventory from ROM files on disk
   - `xml_only` — inventory from an existing `gamelist.xml`
   - `both` — merge both sources (XML entries take priority, new files are appended)

```python
# 1) Scan one platform folder (NES)
nes_df = scan_platform_folder(test_root / 'nes')
print('NES scan:')
display(nes_df)

# 2) Scan all platform folders under root (non-strict: all files are treated as games)
all_df = scan_games_folders(test_root, strict_extensions=False)
print('All platforms (strict_extensions=False):')
display(all_df)

# 3) Strict scan using platform formats
all_strict_df = scan_games_folders(test_root, strict_extensions=True)
print('All platforms (strict_extensions=True):')
display(all_strict_df)

print('Rows (non-strict):', len(all_df))
print('Rows (strict):', len(all_strict_df))
```

## Build / Update `gamelist.xml` — unified pipeline

Test the new `build_platform_gamelist()` with each source mode (`files_only`, `xml_only`, `both`) and the multi-platform `build_all_gamelists()` runner.

```python
from utils.gamelist_builder import build_platform_gamelist, build_all_gamelists, load_metadata

# --- Single platform: files_only (no metadata enrichment) ---
result_files = build_platform_gamelist(
    test_root / 'nes',
    output_path=Path('output/test_games/nes/gamelist_files_only.xml'),
    source='files_only',
)
print(result_files)
```

```python
# --- Single platform: xml_only (uses existing gamelist.xml in the folder) ---
# This requires a gamelist.xml to already exist in the platform folder.
# We'll use the lists/ folder which has gamelist.xml files.
result_xml = build_platform_gamelist(
    Path('lists/nes'),
    output_path=Path('output/test_games/nes/gamelist_xml_only.xml'),
    source='xml_only',
)
print(result_xml)
```

```python
# --- Single platform: both (merge XML + file scan) ---
# The test_root games folder is read-only, so we stage a test folder
# with both a gamelist.xml and some ROM files.
import shutil

both_test = Path('output/test_games/nes_both')
both_test.mkdir(parents=True, exist_ok=True)

# Copy a few ROMs from the scanner results
for _, row in nes_df.head(5).iterrows():
    src = Path(row['source_path'])
    dst = both_test / src.name
    if src.exists() and not dst.exists():
        shutil.copy2(src, dst)

# Copy an existing gamelist.xml into the same folder
gamelist_src = Path('lists/nes/gamelist.xml')
if gamelist_src.exists():
    shutil.copy2(gamelist_src, both_test / 'gamelist.xml')

result_both = build_platform_gamelist(
    both_test,
    output_path=Path('output/test_games/nes_both/gamelist_updated.xml'),
    source='both',
)
print(result_both)
```

```python
# --- Multi-platform runner: process all subfolders under games root ---
results = build_all_gamelists(
    test_root,
    output_root=Path('output/test_games'),
    source='files_only',
)
for r in results:
    print(f"  {r.get('platform_key')}: inventory={r.get('inventory_count')} matched={r.get('matched_count')}")
```

## End-to-end: metadata enrichment from `.pkl`

Load the cleaned metadata dataset and use it to match & enrich gamelist entries.
You can pass `metadata_path=` directly instead of loading into a variable first.

```python
# Load the metadata dataset once
METADATA_PATH = 'output/cleaned_df.pkl'

game_df = load_metadata(METADATA_PATH)
print(f"Metadata loaded: {game_df.shape[0]} games, {game_df['platform'].nunique()} platforms")
display(game_df.head(3))
```

```python
# Single platform with metadata: scan ROMs + match against dataset
result_enriched = build_platform_gamelist(
    test_root / 'nes',
    output_path=Path('output/test_games/nes/gamelist_enriched.xml'),
    game_df=game_df,
    source='files_only',
)
print(f"Inventory: {result_enriched['inventory_count']}, Matched: {result_enriched['matched_count']}, Rate: {result_enriched['match_rate']:.1%}")
```

```python
# Verify enriched XML has metadata fields
import xml.etree.ElementTree as ET

tree = ET.parse('output/test_games/nes/gamelist_enriched.xml')
root = tree.getroot()
games = root.findall('game')
print(f"Total game entries: {len(games)}")

# Show a sample enriched entry
for game in games[:5]:
    name = game.findtext('name', '?')
    desc = game.findtext('desc', '')
    dev = game.findtext('developer', '')
    pub = game.findtext('publisher', '')
    print(f"  {name[:40]:40s}  dev={dev[:20]:20s}  pub={pub[:20]:20s}  desc={'yes' if desc else 'no'}")
```

```python
# Alternative: pass metadata_path directly (loads pkl internally)
# Useful when you don't need the DataFrame in your notebook scope.
result_via_path = build_platform_gamelist(
    Path('lists/nes'),
    output_path=Path('output/test_games/nes/gamelist_xml_enriched.xml'),
    metadata_path=METADATA_PATH,
    source='xml_only',
)
print(f"Inventory: {result_via_path['inventory_count']}, Matched: {result_via_path['matched_count']}, Rate: {result_via_path['match_rate']:.1%}")
```

```python
# Multi-platform with metadata: process selected platforms
results_enriched = build_all_gamelists(
    test_root,
    output_root=Path('output/test_games'),
    platform_keys=['nes', 'gb', 'genesis'],
    game_df=game_df,
    source='files_only',
)
for r in results_enriched:
    status = 'OK' if r.get('success') else 'FAIL'
    print(f"  [{status}] {r.get('platform_key'):12s}  inventory={r.get('inventory_count', 0):5d}  matched={r.get('matched_count', 0):5d}  rate={r.get('match_rate', 0):.1%}")
```
