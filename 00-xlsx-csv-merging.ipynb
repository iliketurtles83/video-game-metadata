### Excel and CSV exploration and merging

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# import xlsx/Game.xlsx
game_xlsx = pd.read_excel('xlsx/Game.xlsx')
print(game_xlsx.shape[0])
game_xlsx.head()
```

```python
game1_xlsx = pd.read_excel('xlsx/Game1.xlsx')
print(game1_xlsx.shape[0])
game1_xlsx.head()
```

```python
# merge game_xlsx and game1_xlsx
game_merged = pd.concat([game_xlsx, game1_xlsx], ignore_index=True).drop_duplicates(ignore_index=True)
print(game_merged.shape[0])
game_merged.head()
```

```python
# Simple check of the data structure first
print("Shape of game_merged:", game_merged.shape)
print("\nColumn names:")
print(game_merged.columns.tolist())
print("\nFirst few rows:")
print(game_merged.head())
```

```python
# Visualize missing vs non-missing data percentages for each column
import numpy as np

# Calculate missing and non-missing percentages for each column
missing_percentages = (game_merged.isna().sum() / len(game_merged)) * 100
non_missing_percentages = 100 - missing_percentages

# Create the visualization
fig, ax = plt.subplots(figsize=(12, 8))

# Create stacked horizontal bar chart
columns = game_merged.columns
y_positions = np.arange(len(columns))

# Plot non-missing data (blue) and missing data (red)
bars_non_missing = ax.barh(y_positions, non_missing_percentages, 
                          label='Non-Missing Data', color='lightblue', alpha=0.8)
bars_missing = ax.barh(y_positions, missing_percentages, 
                      left=non_missing_percentages, label='Missing Data', 
                      color='lightcoral', alpha=0.8)

# Customize the plot
ax.set_yticks(y_positions)
ax.set_yticklabels(columns)
ax.set_xlabel('Percentage (%)')
ax.set_title('Missing vs Non-Missing Data Percentage by Column', fontsize=14, fontweight='bold')
ax.legend()
ax.grid(axis='x', alpha=0.3)

plt.tight_layout()
plt.show()

print(f"\nData Summary:")
print(f"Total rows: {len(game_merged)}")
print(f"Total columns: {len(columns)}")
print(f"Columns with missing data: {(missing_percentages > 0).sum()}")
```

```python
# Create a focused view on just missing data percentages
fig, ax = plt.subplots(figsize=(10, 8))

# Sort columns by missing percentage for better visualization
sorted_missing = missing_percentages.sort_values(ascending=True)

# Create horizontal bar chart for missing data only
bars = ax.barh(range(len(sorted_missing)), sorted_missing.values, 
               color='lightcoral', alpha=0.8)

# Customize the plot
ax.set_yticks(range(len(sorted_missing)))
ax.set_yticklabels(sorted_missing.index)
ax.set_xlabel('Missing Data Percentage (%)')
ax.set_title('Missing Data Percentage by Column (Sorted)', fontsize=14, fontweight='bold')
ax.grid(axis='x', alpha=0.3)

# Add percentage labels on bars
for i, (col, pct) in enumerate(sorted_missing.items()):
    if pct > 0.5:  # Only show label if there's any meaningful missing data
        ax.text(pct + 1, i, f'{pct:.1f}%', va='center', ha='left', fontweight='bold')
    elif pct > 0:
        ax.text(pct + 0.2, i, f'{pct:.1f}%', va='center', ha='left', fontweight='bold', fontsize=10)

plt.tight_layout()
plt.show()

# Print detailed missing data statistics
print("\nDetailed Missing Data Report:")
print("=" * 50)
for col in sorted_missing.index:
    missing_count = game_merged[col].isna().sum()
    missing_pct = (missing_count / len(game_merged)) * 100
    print(f"{col:20}: {missing_count:6,} missing ({missing_pct:5.2f}%)")
```
